```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd

url = "https://resumendiario.elturf.com/"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')


```

```python
data = soup.find_all('td', class_='ttt_estad_estad_td')

# Extract the data from the HTML
extracted_data = []
for item in data:
    extracted_data.append(item.text)

# Organize the data into a tabular format. You could use pandas for this.
df = pd.DataFrame(extracted_data)
```

```python
with open('extracted_data_og.txt', 'w') as file:
    for item in extracted_data:
        file.write(item + '\n')
```

```python
# Define the column names
column_names = ["nro_carrera", "tipo_carrera", "tipo_carrera_alt", "distancia", "lugar_1", "lugar_2", "lugar_3", "empty"]
# Use the index of '\xa0' to split the extracted_data into two lists
index = extracted_data.index('\xa0')

# The first list is from the start to the index of '\xa0'
extracted_data1 = extracted_data[:index]
extracted_data2 = extracted_data[index:]

# Break the extracted_data_1 into rows every 8 columns
rows1 = [extracted_data1[i:i+8] for i in range(0, len(extracted_data1), 8)]

# Create a new dataframe with the rows and column names
df = pd.DataFrame(rows1, columns=column_names)
orig_df = df

# Split the data in 'lugar_1', 'lugar_2', and 'lugar_3' columns
# Use regular expression to remove data after ")"
df['lugar_1'] = df['lugar_1'].apply(lambda x: re.sub(r'\).*', ')', x))
df['lugar_2'] = df['lugar_2'].apply(lambda x: re.sub(r'\).*', ')', x))
df = df.drop('empty', axis=1)
```

```python
df['lugar_1'] = df['lugar_1'].str.slice(3)
df['lugar_2'] = df['lugar_2'].str.slice(3)
df['lugar_3'] = df['lugar_3'].str.slice(3)

```

```python
df['lugar_1_semental'] = df['lugar_1'].str.extract(r'\((.*?)\)')
df['lugar_1'] = df['lugar_1'].str.replace(r' \(.*?\)', '', regex=True)

df['lugar_2_semental'] = df['lugar_2'].str.extract(r'\((.*?)\)')
df['lugar_2'] = df['lugar_2'].str.replace(r' \(.*?\)', '', regex=True)
```
